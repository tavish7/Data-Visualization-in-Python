```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline # helps to include graphs inside jupyter notebook instead of showing up as a pop-up window
```

### Line Chart

```python
yield_apples = [0.23, 1.52, 3.5, 3.8, 4.5, 5.5, 6]
#plt.plot(yield_apples)

# We can use ; at the end of the plt.plot() function to avoid printing output result and only show the graph
plt.plot(yield_apples);
```

Adding X-Axis

```python
years = [2000, 2002, 2004, 2006, 2008, 2012, 2014]
plt.plot(years, yield_apples);
```

Axis Label

```python
plt.plot(years, yield_apples)
plt.xlabel("Years")
plt.ylabel("Yield in tons per hectare");
```

Multiple lines

```python
orange = [9, 8.5, 6.5 ,5.9, 3.2, 2.6, 1.2]
plt.plot(years, yield_apples)
plt.plot(years, orange)
plt.xlabel("Years")
plt.ylabel("Yield in tons per hectare");
```
