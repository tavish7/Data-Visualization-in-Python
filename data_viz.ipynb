```python
import matplotlib.pyplot as plt
import seaborn as sns

#helps to include graphs inside jupyter notebook instead of showing up as a pop-up window
%matplotlib inline 
```

### Line Chart

```python
yield_apples = [0.23, 1.52, 3.5, 3.8, 4.5, 5.5, 6]
#plt.plot(yield_apples)

# We can use ; at the end of the plt.plot() function to avoid printing output result and only show the graph
plt.plot(yield_apples);
```

Adding X-Axis

```python
years = [2000, 2002, 2004, 2006, 2008, 2012, 2014]
plt.plot(years, yield_apples);
```

Axis Label

```python
plt.plot(years, yield_apples)
plt.xlabel("Years")
plt.ylabel("Yield in tons per hectare");
```

Multiple lines

```python
orange = [9, 8.5, 6.5 ,5.9, 3.2, 2.6, 1.2]
plt.plot(years, yield_apples)
plt.plot(years, orange)
plt.xlabel("Years")
plt.ylabel("Yield in tons per hectare");
```

Legends and Chart Titles

```python
plt.plot(years, yield_apples)
plt.plot(years, orange)

plt.xlabel("Years")
plt.ylabel("Yield in tons per hectare")

plt.title("Crop Yields")
plt.legend(['Apples', 'Oranges'])
```

Line Marker

```python
plt.plot(years, yield_apples, marker = '*')
plt.plot(years, orange, marker = 'o')

plt.xlabel("Years")
plt.ylabel("Yield in tons per hectare")

plt.title("Crop Yields")
plt.legend(['Apples', 'Oranges'])
```

Styling Markers

```python
plt.plot(years, yield_apples, marker='*', c='y', ls='-', lw=1, ms=7, mew=3, mec='navy')
plt.plot(years, orange, marker='o', c='g', ls='--', lw=3, ms=5, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Yield in tons per hectare')

plt.title("Crop Yields")
plt.legend(['Apples', 'Oranges']);
```

`fmt` is used to provide shorthand for markers

```python
fmt = '[marker][line][color]'

plt.plot(years, yield_apples, 's-b')
plt.plot(years, orange, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield in tons per hectare')

plt.title("Crop Yields")
plt.legend(['Apples', 'Oranges']);
```

Figure Size

```python
plt.figure(figsize=(10,5))

plt.plot(years, yield_apples, 's-b')
plt.plot(years, orange, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield in tons per hectare')

plt.title("Crop Yields")
plt.legend(['Apples', 'Oranges']);
```

## Scatter Plot

```python
# Load data
flower = sns.load_dataset('iris')
flower
```

```python
flower.species.unique()
```

```python
sns.scatterplot(x = flower.sepal_length, y = flower.sepal_width);
```

We can add `hue` to differentiate based on particular attribute and `s` to adjust the point size

```python
sns.scatterplot(x = flower.sepal_length, y = flower.sepal_width, hue = flower.species, s = 50);
```

```python
plt.Figure(figsize=(10,7))
plt.title("Sepal Dimension")

sns.scatterplot(x = flower.sepal_length, 
                y = flower.sepal_width,
                hue = flower.species,
                s = 50);
```

## Histogram

```python
plt.title("Distribution")

sns.histplot(flower.sepal_width);
```

```python
plt.hist(flower.sepal_width);
```

```python
# Specifying bins
plt.hist(flower.sepal_width, bins = 7);
```

## Multiple Histogram

```python
import numpy as np

setosa = flower[flower.species == 'setosa']
versicolor = flower[flower.species == 'versicolor']
virginica = flower[flower.species == 'virginica']

plt.hist(setosa.sepal_width, alpha = 0.5, bins = np.arange(2,5, 0.25));
plt.hist(versicolor.sepal_width, alpha = 0.5, bins = np.arange(2,5, 0.25));
plt.hist(virginica.sepal_width, alpha = 0.5, bins = np.arange(2,5, 0.25));
```

```python
plt.title('Distribution of Sepal Width')

plt.hist([setosa.sepal_width, versicolor.sepal_width, virginica.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=False);

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

```python
plt.title('Distribution of Sepal Width')

plt.hist([setosa.sepal_width, versicolor.sepal_width, virginica.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```
